### Q2 [60 Points, M]
Given the PCA code bellow, 
* annotate the axes acoording to the PCA component number (i.e., the axis corresponding to the first principal component (PC) should be labeled as First PC and second axis as Second PC and so on)
* find the corresponding eigenvalue to each PC and place inside the axis label (e.g., First PC with $\lambda_1=0.5$)
* draw the chosen three principanl vectors on the 3D plot (hint: the dimension corresponding to the lowest PC should be ignored since PC vectors are 4 dimensional and can not be ploted in a 3D plot)

```
#@markdown
@widgets.interact
def plot_pca_3d(elev=(28, 68, 1), azim=(114, 154, 1)):

  iris = datasets.load_iris()
  X = iris.data
  y = iris.target

  fig = plt.figure(1, figsize=(8, 6))
  plt.clf()
  ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=elev, azim=azim)

  plt.cla()
  pca = decomposition.PCA(n_components=3)
  pca.fit(X)
  X = pca.transform(X)

  for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
      ax.text3D(X[y == label, 0].mean(),
                X[y == label, 1].mean() + 1.5,
                X[y == label, 2].mean(), name,
                horizontalalignment='center',
                bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
  # Reorder the labels to have colors matching the cluster results
  y = np.choose(y, [1, 2, 0]).astype(np.float)
  ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)

  ax.w_xaxis.set_ticklabels([])
  ax.w_yaxis.set_ticklabels([])
  ax.w_zaxis.set_ticklabels([])

  plt.show()
```
