### Q3 [10 Points, S]
Prove that if there is a zero eigenvalue, the matrix is singular.

Solution:

By definition: $Av = \lambda v$

If $\lambda=0$, then $Av=0$ and by definition $v\neq0$

Since this is a homogenous system, A has infinitely many solutions which makes A non-invertable and, in turn, a singular matrix.

### Q4 [10 Points, M]
Given $f(\boldsymbol{x})=\boldsymbol{x}^{\top} \boldsymbol{A} \boldsymbol{x}$ and $\|\boldsymbol{x}\|_{2}=1$, prove that Whenever $x$ is equal to an eigenvector of $A$, $f$ takes on the value of the corresponding eigenvalue

Solution:

$x = v$ where $v$ is the eigenvector of $A$

$f(v) = v^TAv$

$= v^T\lambda v = \lambda v v^T$

$= \lambda ||v||^2_2$

Since the norm is 1, $f(v) = \lambda$

### Q5 [10 Points, M]
Prove that positive definite matrices guarantee that $x^{\top} A x=0 \Rightarrow x=0$

Solution:

We first need to show that $x$ is eigenvector of $A$.

Definition of positive semidefinite matrices: $x^TAx \geq 0$

As a result, $x^TAx = 0$ corresponds to minimum value of the $x^TAx$.

It is known that minimum value of $x^TAx$ is equal to minimum eigenvalue of $A$. Then, $x$ must be the eigenvector of $A$.

Now, we can write the statement like $v^TAv = 0$

$=v^T \lambda v = \lambda ||v||_2^2 = 0$

By definition of positive definite matrices, $\lambda>0$. So, $||v||_2^2 = 0$, and to hold that, $v$ must be a zero vector.

### Q6 [10 Points, H]
Prove that if $\lambda$ is an eigenvalue of $A$ then $\lambda^n$ is an eigenvalue of $A^n$

Solution:

We will use proof by induction.

We assume that $n=k$ and $A^kv = \lambda^kv$

Now, we will show it is true for $n = k+1$

$A A^kv = A \lambda^kv$

$ = A^{k+1}v = \lambda^kAv$

$ = A^{k+1}v = \lambda^k \lambda v$

$ = A^{k+1}v = \lambda^{k+1} v$

### Q7 [20 Points, H]
Given a fully connected neural network defined as,
$$
\mathbf{v}_{\text {out}}=\mathbf{A_1} \cdot \mathbf{A_2} \cdot \cdots \mathbf{A_n} \mathbf{v}_{\text {in}}
$$
Prove that if eignevalues of all the weights are equal to each other and they are less than 1, the network output for any input converges to zero as network gets deeper, i.e., $n \rightarrow \infty$. 

Also, argue the importance of weight initialization using this question.

Solution:

First, we will show that the property that we proved at the previous question holds for different $A$s.

Assume that $n=k$ and $A_k.A_{k-1}...A_1 v = \lambda^k v$

Show that it is true for $n=k+1$

 $A_{k+1}.A_k.A_{k-1}...A_1 v = A_{k+1}\lambda^k v$

 $= \lambda v \lambda^k$

  $= \lambda^{k+1}v$

  We can write $v_{out} = A. v_{in}$ where $A=A_1.A_2...A_n$

  $v_{out} = Q\Lambda Q^T. v_{in}$

  We showed that $\Lambda$ is composed of $\lambda^n$. 

  $\lim_{n \to \infty} \lambda^n = 0$

  Therefore, $v_{out} =0$ regardless of $v_{in}$

  This example shows the importance of weight initialization because if you initialize the weights very small values like here, your network will always return 0 as output and won't learn anything.

### Q9 [10 Points, M]
Prove that the right-singular vectors of $A$ are the eigenvectors of $A^TA$

Solution:

Eigendecomposition of $A^TA= (Q\Lambda Q^T)(Q\Lambda Q^T)^T$

$=Q\Lambda Q^T Q\Lambda^T Q^T$

$=Q\Lambda I \Lambda^T Q^T$

$=Q\Lambda \Lambda^T Q^T$

So, $Q\Lambda \Lambda^T Q^T = UDV^T$

Then, we can see that 
$Q = U, \Lambda \Lambda^T = D, Q^T = V^T$
